```python
import requests
import csv

url = 'https://jsonplaceholder.typicode.com/users'

response = requests.get(url)

if response.status_code == 200:
    users = response.json() 
    
    csv_columns = [
        'id', 'name', 'username', 'email', 'street', 'suite', 'city', 'zipcode', 
        'lat', 'lng', 'phone', 'website', 'company_name', 'catchPhrase', 'bs'
    ]
    
    with open('users_data.csv', 'w', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader() 
        
        for user in users:
            writer.writerow({
                'id': user['id'],
                'name': user['name'],
                'username': user['username'],
                'email': user['email'],
                'street': user['address']['street'],
                'suite': user['address']['suite'],
                'city': user['address']['city'],
                'zipcode': user['address']['zipcode'],
                'lat': user['address']['geo']['lat'],
                'lng': user['address']['geo']['lng'],
                'phone': user['phone'],
                'website': user['website'],
                'company_name': user['company']['name'],
                'catchPhrase': user['company']['catchPhrase'],
                'bs': user['company']['bs']
            })
    
    print("Data has been successfully extracted and saved in 'users_data.csv'.")
else:
    print(f"Failed to connect to the API. Error code: {response.status_code}")
```

```python
import requests as req
from bs4 import BeautifulSoup
import csv

req = req.get('https://www.scrapethissite.com/pages/forms/?page_num=1')
soup = BeautifulSoup(req.content, 'html.parser')
teams = soup.find_all('tr', attrs={'class': 'team'})

name = [team.find('td', {'class': 'name'}).text.strip() for team in teams]
year = [team.find('td', {'class': 'year'}).text.strip() for team in teams]
wins = [team.find('td', {'class': 'wins'}).text.strip() for team in teams]
losses = [team.find('td', {'class': 'losses'}).text.strip() for team in teams]
ot_losses = [team.find('td', {'class': 'ot-losses'}).text.strip() for team in teams]
win_pct = [team.find('td', {'class': 'pct'}).text.strip() for team in teams]
gf = [team.find('td', {'class': 'gf'}).text.strip() for team in teams]
ga = [team.find('td', {'class': 'ga'}).text.strip() for team in teams]

with open('teams_data.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Name', 'Year', 'Wins', 'Losses', 'OT Losses', 'Win', 'GF', 'GA'])
    for i in range(len(teams)):
        writer.writerow([name[i], year[i], wins[i], losses[i], ot_losses[i], win_pct[i], gf[i], ga[i]])

print("Data has been saved to the file teams_data.csv")
```
